### Аномалии в данных

**При осмотре данных было замечено, что некоторые параметры имеют разброс на несколько порядков или "странные" отрицательные значения**

Отбросим 1% экстремальных значений в столбцах, в которых выполнено хотя бы одно из условий:
- максимум более чем в 100 раз превышает значение 99,5% квантиля
- минимум меньше 0, а значение 0,5% квантиля больше 0


```python
problem_column_train = []
problem_column_test = []
problem_column_full = []

for df in df_list:
    print(df.name, '\n')
    for column in df.columns:
        if column != 'date' and ((df[column].max() / df[column].quantile(.995) > 100) \
        or (df[column].min() * df[column].quantile(.005)) < 0):
            if df.name == 'train':
                problem_column_train.append(column)
            elif df.name == 'test':
                problem_column_test.append(column)
            else:
                problem_column_full.append(column)
            print('\n', column)
            print(df[column].describe())
```

**Действительно, выбраны столбцы, в которых либо есть выбросы, отличающиеся на порядки от медианы, либо  есть абсурдные значения (отрицательный объем воздуха)**

**Избавимся от этих данных**

**Также, можно предположить, что околонулевые концентрации элементов или размеры частиц в исходном сырье или на выходе той или иной стадии технологического процесса - это также аномалии**

Посмотрим и на них тоже. При этом околонулевые содержания реактивов аномалиями не считаем: действительно, возможна ситуация, когда при очистке тот или иной реактив просто не используется.

```python
zero_column_train = []
zero_column_test = []
zero_column_full = []

for df in df_list:
    print(df.name, '\n')
    for column in df.columns:
        if 'output.concentrate' in column or 'input.feed' in column:
            print(f'Количество нулевых значений в столбце {column} - {df.loc[df[column] <= 1][column].count()}')
            if df.name == 'train':
                zero_column_train.append(column)
            elif df.name == 'test':
                zero_column_test.append(column)
            else:
                zero_column_full.append(column)
    print('\n')
```

```python
for column in problem_column_full:
    q_995 = df_full[column].quantile(.995)
    q_005 = df_full[column].quantile(.005)
    df_full = df_full.drop(df_full[(df_full[column] > q_995)].index)
    df_full = df_full.drop(df_full[(df_full[column] < q_005)].index)
```

```python
for column in problem_column_train:
    q_995 = df_train[column].quantile(.995)
    q_005 = df_train[column].quantile(.005)
    df_train = df_train.drop(df_train[(df_train[column] > q_995)].index)
    df_train = df_train.drop(df_train[(df_train[column] < q_005)].index)
```

```python
for column in problem_column_test:
    if column in df_test.columns:
        q_995 = df_test[column].quantile(.995)
        q_005 = df_test[column].quantile(.005)
        df_test = df_test.drop(df_test[(df_test[column] > q_995)].index)
        df_test = df_test.drop(df_test[(df_test[column] < q_005)].index)
```

### Функция sMAPE

```python
def s_mape(target, prediction):
    s = 0
    p = prediction
    t = np.array(target)
    for i in range(len(t)):
        s += abs(t[i] - p[i]) / (1e-3 + abs(t[i]) + abs(p[i]))
    return s * 200 / len(t)

smape = make_scorer(s_mape, greater_is_better=False)

def smape_final(model, features_r, target_r, features_f, target_f, n):
    scores_r = cross_val_score(model, features_r, target_r, scoring=smape, cv=n)
    score_mr = abs(sum(scores_r) / len(scores_r))
    scores_f = cross_val_score(model, features_f, target_f, scoring=smape, cv=n)
    score_mf = abs(sum(scores_f) / len(scores_f))
    f = score_mr * .25 + score_mf * .75
    print('Final sMape-- ', model, '\n--', round(f,2))
    return f
```
