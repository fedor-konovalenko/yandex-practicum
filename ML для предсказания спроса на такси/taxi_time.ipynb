#  Прогнозирование заказов такси

Исходные данные -  исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час и построить модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

## Подготовка

### Сезонность для других временных масштабов

```python
query = {'daily': '"2018-06-06 00:00:00":"2018-06-08 12:00:00"',
         'weekly': '"2018-06-04 00:00:00":"2018-06-11 00:00:00"',
         'monthly': '"2018-06-01 00:00:00":"2018-07-01 00:00:00"'}    
for c in query.keys():
    #print(c)
    
    interval = query[c]
    t1 = interval.split(':"')[0].replace('"', '')
    t2 = interval.split(':"')[1].replace('"', '')
    data = df[datetime.datetime.strptime(t1, "%Y-%m-%d %H:%M:%S"):datetime.datetime.strptime(t2, "%Y-%m-%d %H:%M:%S")]
    data_dc = seasonal_decompose(data['num_orders'])
    
    fig = data_dc.plot(observed=False)
    #plt.grid(True)
    fig.set_size_inches((10, 8))
    fig.suptitle(f'{c} decomposed data')
    plt.xlabel('time')
    plt.ylabel('orders')
    plt.show()
```

### Стационарность временного ряда

На первый взгляд ряд выглядит нестационарным: его среднее значение возрастает. Но это только на первый взгляд

```python
alpha = .05
adf_test = adfuller(df['num_orders'])
if adf_test[1] > alpha:
    print(f'p-value = {adf_test[1]} Ряд нестационарный при уровне значимости {alpha}')
else:
    print(f'p-value = {adf_test[1]} Ряд стационарный при уровне значимости {alpha}')
```

Хм, как-то все [неоднозначно](https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html). [Посмотрим по критическим значениям](https://real-statistics.com/statistics-tables/augmented-dickey-fuller-table/)

```python
for cr in ['1%', '5%', '10%']:
    if adf_test[0] < adf_test[4][cr]:
        print(f'Ряд стационарный при уровне значимости {cr}')
    else:
        print(f'Ряд нестационарный при уровне значимости {cr}')
```

### Выводы

**Сезонность**
- Есть ярко выраженная суточная сезонность. Недельная сезонность также заметна.
- В течение дня: много заказывают вечером, ночью спрос резко падает, что обусловлено расписанием рейсов, графиком работы общественного транспорта и, вероятно, загруженностью дорог
- В середине недели спрос выше, чем на выходных, что вероятно объясняется особенностью данного аэропорта: что много рейсов используется для деловых перелетов

**Стационарность ряда**
- по росту скользящего среднего можно сделать вывод о нестационарности временного ряда
- по графику остатков, в целом, остатки распределены стабильно, однако, есть выбросы, что также говорит о нестационарности
- по критерию Дики-Фуллера с уровнем значимости 0,01 ряд также нестационарный

**Ряд нестационарный**

## Обучение

### Подготовка признаков

**Значимые для прогнозирования признаки**
- день недели
- час
- скользящее среднее (уже есть и уже сдвинуто на 1 шаг для недопущения утечки данных)
- запаздывающее значение, сдвинутое на 24 часа (судя по периоду колебаний сезонности и графику автокорреляции)
- запаздывающее значение, сдвинутое на 1 час (судя графику автокорреляции)

```python
plot_acf(df['num_orders'])
plt.grid()
plt.xlabel('lags')
plt.show()
```

```python
df['hour'] = df.index.hour
df['dow'] = df.index.dayofweek
df['lag24'] = df['num_orders'].shift(24)
df['lag1'] = df['num_orders'].shift(1)
df = df.dropna()
df.head()
```

### Корреляция признаков и таргета

```python
for column in df.columns.drop('num_orders'):
        c = df[column].corr(df['num_orders'])
        print(f'Коэффициент корреляции параметра {column} и количества заказов {c}')
pd.plotting.scatter_matrix(df, alpha=.05, figsize=(10,10))
plt.show()
```

### Разделение на выборки

```python
df_train, df_test = train_test_split(df, shuffle=False, test_size=0.1)
```

### Features & Target

```python
features_train = df_train.drop(['num_orders'], axis=1)
features_test = df_test.drop(['num_orders'], axis=1)
target_train = df_train['num_orders']
target_test = df_test['num_orders']
plt.plot(target_train)
plt.plot(target_test, color='green')
plt.grid()
plt.xlabel('time')
plt.ylabel('orders')
plt.title('train and test splits')
plt.show()
```

### Линейная регрессия

```python
model = LinearRegression()
tscv = TimeSeriesSplit(n_splits=5)
result = abs(cross_val_score(model, features_train, target_train, cv=tscv, scoring='neg_mean_squared_error')) ** .5
print('MEAN RMSE ---', sum(result) / len(result))
```

### CatBoost

**Варьируемые гиперпараметры**
- скорость обучения
- количество решателей
- глубина деревьев

```python
model = CatBoostRegressor(verbose=False)
tscv = TimeSeriesSplit(n_splits=5)

grid = {'learning_rate': [.3, .4, .5],
        'depth': [2, 4, 8, 10],
        'n_estimators': [50, 100, 200]}
grid_search_result = model.grid_search(grid, X=features_train, y=target_train, cv=tscv, plot=True, verbose=False) 
best_cat = CatBoostRegressor(**grid_search_result['params'], random_state=806, verbose=False)
joblib.dump(best_cat, 'best_cat.joblib')
print(grid_search_result['params'])
result = grid_search_result["cv_results"]["test-RMSE-mean"]
print(f'mean RMSE --- {sum(result) / len(result)}')
```

### SARIMAX

**Коэффициенты модели**
(p, d, q) = (1,0,0), [взято значение по умолчанию](https://www.statsmodels.org/dev/generated/statsmodels.tsa.statespace.sarimax.SARIMAX.html)
s = 4 (квартальный период)
Блоки для кросс-валидации формируются в цикле из конца выборки, поскольку по умолчанию SARIMAX с TimeSeriesSplit не работает

```python
rmse = []
valid_percent = .25
valid_size = int(valid_percent * features_train.shape[0])
for cv in range(int(1 / valid_percent - 1)):
    train_end = features_train.shape[0] - (cv + 1) * valid_size
    valid_end = train_end + valid_size
    features_train_cv = features_train[:train_end]
    target_train_cv = target_train[:train_end]
    features_valid_cv = features_train[train_end:valid_end]
    target_valid_cv = target_train[train_end:valid_end]
    model = SARIMAX(target_train_cv, order = (1, 0, 0), seasonal_order = (1, 1, 1, 4))
    result = model.fit()
    predictions = result.predict(1, target_valid_cv.shape[0])
    rmse.append(mean_squared_error(predictions, target_valid_cv) ** .5)
print('RMSE MEAN ---', sum(rmse) / len(rmse))
```

### Вывод

**По результатам обучения лучшее значение целевой метрики RMSE (27.2) - показала модель CatBoostRegressor. Линейная регрессия почти не уступает ей: RMSE=27.5**

## Тестирование

### Проверка лучшей модели на тестовой выборке

```python
model = joblib.load('best_cat.joblib')
model.fit(features_train, target_train)
predictions = model.predict(features_test)
rmse = mean_squared_error(predictions, target_test) ** .5
print(f'TEST RMSE --- {rmse}')
```

```python
pred = pd.Series(predictions, index=target_test.index)

#plt.plot(target_train, alpha=.5)
plt.plot(target_test, color='green', alpha=.5, label='target')
plt.plot(pred, color='red', label='catboost prediction')
plt.legend()
plt.xlabel('time')
plt.ylabel('orders')
plt.title('target VS predictions')
plt.grid()
plt.show()
```

### Проверка лучшей модели на адекватность

```python
dummy_model = DummyRegressor()
dummy_model.fit(features_train, target_train)
predictions = dummy_model.predict(features_test)
result = mean_squared_error(predictions, target_test)
print('Dummy RMSE ---', result ** .5)
```
